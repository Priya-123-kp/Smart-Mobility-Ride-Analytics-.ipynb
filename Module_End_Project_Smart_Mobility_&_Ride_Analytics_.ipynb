Probability Analysis-Probability of Peak-Hour Ride

```python
peak_rides = len(df[df['Ride_Time'] == 'Peak'])
total_rides = len(df)
prob_peak = peak_rides/total_rides
print(prob_peak)
```

Conditional Probability

```python
high_fare = df['Fare_Amount'].median()
cond_prob = len(df[(df['Ride_Time'] == 'Peak') & (df['Fare_Amount'] > high_fare)])/len(df)
print(cond_prob)
```

Detect outliers using percentiles, quartiles, IQR, and z-score

```python
Q1 = df['Fare_Amount'].quantile(0.25)
Q3 = df['Fare_Amount'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['Fare_Amount'] < lower) | (df['Fare_Amount'] > upper)]
print(outliers)
```

Z-Score

```python
df['z_score'] = stats.zscore(df['Fare_Amount'])
outliers = df[np.abs(df['z_score']) > 3]
print(outliers)
```

Measure relationships using Pearson, Spearman, and covariance

Pearson Correlation

```python
corr,pvalue =stats.pearsonr(df['Trip_Distance'],df['Fare_Amount'])
print(corr,pvalue)
```

premium rides statistically more expensive than economy rides
Independent T-Test
Hypothesis

```python
Premium = df[df['Ride_Category'] =="Premium"]['Fare_Amount']
economy = df[df['Ride_Category'] =="Economy"]['Fare_Amount']
t_stat,p_value = stats.ttest_ind(Premium,economy)
print(t_stat,p_value)
```

A relationship between surge pricing and customer ratings

```python
corr,p = stats.spearmanr(df['Surge_Multiplier'],df['Customer_Rating'])
print(corr,p)
```

peak-hour rides differ in revenue compared to non-peak rides

```python
peak = df[df['Ride_Time']=='Peak']['Fare_Amount']
non_peak =df[df['Ride_Time']=='Non-Peak']['Fare_Amount']
stats.ttest_ind(peak,non_peak)
```

```python
mean = df['Fare_Amount'].mean()
std = df['Fare_Amount'].std()
n = len(df)
ci = stats.norm.interval(0.95,loc= mean,scale= std/np.sqrt(n))
print(ci)
```
